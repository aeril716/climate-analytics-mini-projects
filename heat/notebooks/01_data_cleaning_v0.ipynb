```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/aeri/Desktop/Climate Risk/MINIPROJECT/USW00023234.csv', low_memory=False)

# dates
df['DATE'] = pd.to_datetime(df['DATE'])

# necessary columns extract
cols = ['DATE', 'TMAX', 'TMIN', 'TAVG']
df = df[cols]

# TAVG calculation (in case of no data)
df['TAVG'] = df['TAVG']. fillna((df['TMAX'] + df['TMIN']) / 2)

# NOAA Celsius*10 to C
df['TMAX'] = df['TMAX']/10
df['TMIN'] = df['TMIN']/10
df['TAVG'] = df['TAVG']/10

df.to_csv('clean_temperature.csv', index=False)
```
